```python
# URL and Date of Press Release

import numpy as np

url = "http://www.senat.fr/communiques/"
 
# Getting the webpage, creating a Response object.
response = requests.get(url)
 
# Extracting the source code of the page.
data = response.text
 
# Passing the source code to BeautifulSoup to create a BeautifulSoup object for it.
soup = BeautifulSoup(data, "html.parser")
 
# Extracting all the <a> tags into a list.
tags = soup.find_all('a', class_="link")
 
# Extracting URLs and Dates from the attribute href in the <a> tags.

import json

data = {}
arrayOfElements = list()

for tag in tags:
        link = tag.get('href')
        data['link'] = link
        date = tag.get_text()
        data['date'] = date
        arrayOfElements.append(data)

# print(json.dumps(arrayOfElements))

!pip install pandas
import pandas as pnds
web_scraped_dataframe = pnds.DataFrame(arrayOfElements)

print(web_scraped_dataframe.head())
```
